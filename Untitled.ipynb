```python
import pandas as pd
import networkx as nx 
import matplotlib.pyplot as plt
import re

# Above didnt work at first because characters got "_" instead of whitespace meaning references were wrong

df_characters = pd.read_csv("harry_potter_df.csv")
df_characters = df_characters.drop('Unnamed: 0',axis = 1)

for i, (c, s) in enumerate(df_characters[['Character', 'Story']].values):
    # Find all occurences of "[[...]]"
    regex_ref = re.findall('\[\[[^\]\]]*\]\]', s)
    # Remove "[" and "]" from elements in list
    regex_ref = [r.replace('[', '').replace(']', '') for r in regex_ref]
    # Remove duplicates
    references = list(dict.fromkeys(regex_ref))
  
    # Remove all references not prominent in the superhero list
    references = [ref for ref in references if len(df_characters[df_characters['Character'] == ref.replace(' ', '_')])] 
  
    df_characters.loc[i,'References'] = references
df_characters
```

```python
type(df_characters["References"].values[0])
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(df_characters["Character"])
```

```python
print(len(G.nodes))
```

```python
for c,r in df_characters[["Character","References"]].values:
     for single in r:
        if single != c and single in df_characters["Character"].values:
            G.add_edge(c,single)
```

```python
print(len(G.nodes))
print(len(G.edges))
```

```python
df_characters["Character"]
```

```python
nx.draw(G)
```

```python
pos = nx.kamada_kawai_layout(G)
```

```python
from fa2 import ForceAtlas2
forceatlas2 = ForceAtlas2(
                        # Behavior alternatives
                        outboundAttractionDistribution=True,  # Dissuade hubs
                        linLogMode=False,  # NOT IMPLEMENTED
                        adjustSizes=False,  # Prevent overlap (NOT IMPLEMENTED)
                        edgeWeightInfluence=1,

                        # Performance
                        jitterTolerance=0.1,  # Tolerance
                        barnesHutOptimize=True,
                        barnesHutTheta=0.5,
                        multiThreaded=False,  # NOT IMPLEMENTED

                        # Tuning
                        scalingRatio=2.0,
                        strongGravityMode=False,
                        gravity=0.1,

                        # Log
                        verbose=True)

positions = forceatlas2.forceatlas2_networkx_layout(G, pos=pos, iterations=2000)
```

```python
d = dict(G.degree())
plt.figure(1, figsize=(20,12)) 

nx.draw_networkx_nodes(G, pos=positions, node_size=[v * 10 for v in d.values()], label=False)
nx.draw_networkx_edges(G, positions, edge_color="black", alpha=0.05)

plt.title("The ForceAtlas representation of the Character Network.")

plt.show()
```

```python
plt.figure(figsize=(14,7)) 
plt.style.use('seaborn-whitegrid') 
degrees = [G.degree(n) for n in G.nodes()]
plt.hist(degrees, bins=58)
plt.title('Degree distribution for characters') 
plt.xlabel('Degree') 
plt.ylabel('Amount of characters') 
plt.show()
```
